# Mushroom Mania!

In this project we use Seaborn to visualize what features of mushrooms might prove most helpful in building a model to determine whether a given mushroom is edible or poisonous. Note this project is not the model itself. This project showcases the power of Seaborn as a data visualization tool.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# load in the data
df = pd.read_csv("mushroom_data.csv")
#print(df.head())

# list of all column headers
columns = df.columns.tolist()
print(columns)
for i in range(len(columns)):
#  print('Column name: ', columns[i])
  sns.countplot(x = columns[i], data = df, order=df[columns[i]].value_counts().index)
# rotates the value labels slightly so they don’t overlap, also slightly increases font size
  plt.xticks(rotation=30, fontsize=10)
# increases the variable label font size slightly to increase readability
  plt.xlabel(i, fontsize=12)
  plt.title(columns[i] + ' Value Counts')
  plt.show()
  plt.clf()
```
